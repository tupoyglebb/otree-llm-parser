````python
# Построение анкеты, готовой к отправке в LLM
from otree_parser import parse_otreezip, build_questionnaire_for_api
import warnings
warnings.filterwarnings('ignore')

print("=" * 70)
print("ПОСТРОЕНИЕ АНКЕТЫ ДЛЯ ОТПРАВКИ В LLM")
print("=" * 70)

# Парсим эксперимент
experiment = parse_otreezip('otreezips/1factor.otreezip')

# Строим анкеты со всеми вариантами treatment в формате для LLM
questionnaires = build_questionnaire_for_api(
    experiment, 
    include_all_treatment_variants=True, 
    format_for_llm=True
)

print(f"\n✓ Создано анкет: {len(questionnaires)}")
print(f"✓ Формат: список строк, готовых для отправки в LLM")

# Показываем структуру каждой анкеты
for i, questionnaire in enumerate(questionnaires, 1):
    print(f"\n{'='*70}")
    print(f"АНКЕТА #{i}")
    print(f"{'='*70}")
    print(f"Количество элементов: {len(questionnaire)}")
    
    # Показываем первые 3 элемента
    print(f"\nПервые 3 элемента:")
    for j, item in enumerate(questionnaire[:3], 1):
        print(f"\n  Элемент {j}:")
        # Показываем первые 200 символов
        preview = item[:200] + "..." if len(item) > 200 else item
        print(f"  {preview}")
    
    if len(questionnaire) > 3:
        print(f"\n  ... и еще {len(questionnaire) - 3} элементов")
````

````python
# Статистика по всем анкетам
print("=" * 70)
print("СТАТИСТИКА ПО ВСЕМ АНКЕТАМ")
print("=" * 70)

total_elements = sum(len(q) for q in questionnaires)
avg_elements = total_elements / len(questionnaires) if questionnaires else 0

print(f"\n📊 Общая статистика:")
print(f"   • Всего анкет: {len(questionnaires)}")
print(f"   • Всего элементов: {total_elements}")
print(f"   • Среднее элементов на анкету: {avg_elements:.1f}")

print(f"\n📋 Детали по каждой анкете:")
for i, questionnaire in enumerate(questionnaires, 1):
    # Подсчитываем типы элементов
    treatment_info_count = sum(1 for item in questionnaire if item.startswith("ИНФОРМАЦИЯ О TREATMENT"))
    instructions_count = sum(1 for item in questionnaire if item.startswith("ИНСТРУКЦИЯ"))
    questions_count = sum(1 for item in questionnaire if item.startswith("ВОПРОС"))
    
    print(f"\n   Анкета #{i}:")
    print(f"      • Всего элементов: {len(questionnaire)}")
    print(f"      • Информация о treatment: {treatment_info_count}")
    print(f"      • Инструкции: {instructions_count}")
    print(f"      • Вопросы: {questions_count}")
````

````python
# Сохранение анкет в файлы (опционально)
import json
from pathlib import Path

print("=" * 70)
print("СОХРАНЕНИЕ АНКЕТ В ФАЙЛЫ")
print("=" * 70)

output_dir = Path("questionnaires_output")
output_dir.mkdir(exist_ok=True)

for i, questionnaire in enumerate(questionnaires, 1):
    # Сохраняем как JSON
    json_file = output_dir / f"questionnaire_{i}.json"
    with open(json_file, 'w', encoding='utf-8') as f:
        json.dump(questionnaire, f, ensure_ascii=False, indent=2)
    
    # Сохраняем как текстовый файл (для удобного чтения)
    txt_file = output_dir / f"questionnaire_{i}.txt"
    with open(txt_file, 'w', encoding='utf-8') as f:
        f.write("=" * 70 + "\n")
        f.write(f"АНКЕТА #{i}\n")
        f.write("=" * 70 + "\n\n")
        for j, element in enumerate(questionnaire, 1):
            f.write(f"{'─'*70}\n")
            f.write(f"ЭЛЕМЕНТ {j}\n")
            f.write(f"{'─'*70}\n")
            f.write(element)
            f.write("\n\n")
    
    print(f"✓ Анкета #{i} сохранена:")
    print(f"   • JSON: {json_file}")
    print(f"   • TXT: {txt_file}")

print(f"\n✓ Все анкеты сохранены в директорию: {output_dir}")
````

````python
# Функция для получения анкеты по номеру treatment варианта
def get_questionnaire_by_treatment(questionnaires: list, treatment_number: int):
    """
    Получает анкету по номеру treatment варианта.
    
    Args:
        questionnaires: список анкет из build_questionnaire_for_api()
        treatment_number: номер treatment варианта (начиная с 1)
    
    Returns:
        Анкета (список строк) или None, если номер неверный
    """
    if 1 <= treatment_number <= len(questionnaires):
        return questionnaires[treatment_number - 1]
    return None

# Пример использования
print("=" * 70)
print("ПОЛУЧЕНИЕ КОНКРЕТНОЙ АНКЕТЫ")
print("=" * 70)

treatment_num = 2
selected_questionnaire = get_questionnaire_by_treatment(questionnaires, treatment_num)

if selected_questionnaire:
    print(f"\n✓ Получена анкета для treatment варианта #{treatment_num}")
    print(f"   Количество элементов: {len(selected_questionnaire)}")
    print(f"\n   Первый элемент (treatment info):")
    print(f"   {selected_questionnaire[0]}...")
    print(f"\n   Второй элемент (инструкции):")
    if len(selected_questionnaire) > 1:
        print(f"   {selected_questionnaire[1]}...")
else:
    print(f"\n❌ Treatment вариант #{treatment_num} не найден")
````

## Резюме

Этот ноутбук демонстрирует построение анкет, готовых к отправке в LLM:

1. **Парсинг эксперимента** - извлечение структуры из `.otreezip` файла
2. **Построение анкет** - создание отдельных анкет для каждого treatment варианта
3. **Форматирование для LLM** - каждый элемент анкеты - это готовая строка для отправки в API
4. **Структура анкеты**:
   - Элемент 1: Информация о treatment (если есть)
   - Элемент 2: Инструкции (собранные со всех страниц)
   - Элементы 3+: Вопросы с вариантами ответов

### Использование:

```python
# Получить все анкеты
questionnaires = build_questionnaire_for_api(
    experiment, 
    include_all_treatment_variants=True, 
    format_for_llm=True
)

# Отправить каждый элемент в LLM
for questionnaire in questionnaires:
    for element in questionnaire:
        response = send_to_llm_api(element)
        # Обработать ответ
```
